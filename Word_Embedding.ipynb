```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences= ['Hello How is the weather today','Why MJ whent for baseball',
           'MJ or Lebron the Greatest of all time','What are the current affairs',
           'working on natural language processing','i am a good developer']
```

```python
sentences
```

```python
# vocabulary size
voc_size=10000
```

```python
one_hot_repr=  [one_hot(words,voc_size)for words in sentences]
print(one_hot_repr)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sent_length=8
pad_sent= pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(pad_sent)
```

```python
dim= 15
```

```python
model= Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(pad_sent))
```
